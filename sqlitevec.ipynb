```python
import json
```

```python
def load_documents(file_path):
    """Load documents from a JSON file"""
    with open(file_path, 'r', encoding='utf-8') as file:
            documents = json.load(file)
    print(f"Successfully loaded {len(documents)} documents from {file_path}")
    return documents
```

```python
documents = load_documents("Surah123.json")
```

```python
# Prepare texts and metadata for insertion
texts = [doc["Translation"] for doc in documents]
metadatas = [
    {
        "ID": doc["ID"],
        "Tag1": doc["Tag1"], 
        "Tag2": doc["Tag2"]
    } 
    for doc in documents
]
```

```python
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
embedding_function = SentenceTransformerEmbeddings(model_name="BAAI/bge-large-en")
```

```python
from langchain_community.vectorstores import SQLiteVec
```

```python
connection = SQLiteVec.create_connection(db_file="quraan_vectors.db")
```

```python
vector_store = SQLiteVec(
    table="quraan_collection", embedding=embedding_function, connection=connection
)
```

```python
vector_store.add_texts(texts=texts, metadatas=metadatas)
```

```python
# List of queries
queries = [
    "What is the waiting period (‘iddah) for a divorced woman?",
    "I want to help a struggling relative, but I don’t know if giving them zakat is allowed."
    "Is zakat due on gold jewelry that is not worn regularly?",
    "What are the Islamic rulings on adoption, and does the child inherit from adoptive parents?",
    "Can a Muslim woman marry a non-Muslim man who is from Ahl al-Kitab?",
    "I'm thinking of buying a car on installment with interest.",
    "I signed up for an insurance plan — is that considered haram or a necessity?",
    "How should inheritance be distributed among siblings in Islam?",
    "What does the Qur'an say about helping the poor and needy?",
    "I work at a bank where interest (riba) is involved.",
    "what should a person do if they miss a fast during Ramadan due to illness or travel?",
    "What is the Qur'anic view on giving false testimony or lying in court?",
    "I’ve made a lot of mistakes in life and now I want to return to Islam. What does the Qur'an say about repentance and forgiveness?",
    "I gave someone a loan, and they’re refusing to return it. ",
    "I’ve divorced my wife in anger but now we both want to reconcile. Is there anything in the Qur'an about taking her back?",
    "I want to marry someone from another faith who respects Islam. What guidance does the Qur'an give in such cases?",
    "I had to take a loan with interest due to financial hardship—does the Qur'an say anything about exceptions?",
]

# Batch process each query
for query in queries:
    print(f"Query: {query}")
    results = vector_store.similarity_search(query, k=3)

    for i, result in enumerate(results):
        print(f"  Result {i+1}:")
        print(f"    Translation: {result.page_content}")
        if result.metadata:
            print(f"    ID: {result.metadata.get('ID', 'N/A')}")
            print(f"    Tag1: {result.metadata.get('Tag1', 'N/A')}")
            print(f"    Tag2: {result.metadata.get('Tag2', 'N/A')}")
    print("-" * 80)
```
